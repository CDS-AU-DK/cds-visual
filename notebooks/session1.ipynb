## Image translation

Reload the original image

````python
path_to_image = os.path.join("..", "data", "img", "trex.png")
````

````python
image = cv2.imread(path_to_image)
````

````python
jimshow(image, "Original")
````

Translation is the shifting of an image along the x and y axis. Using translation, we can shift an image up, down, left, or right, along with any combination of the above.

If we know the amount that we want to shift (tx,ty), then we can define the traslation matrix M like this:

<img src="../data/sample_data/translate.png">

__Shift image down and right__

We want to shift the whole image _right_ by 25 pixels and _down_ by 50.

````python
M = np.float32([[1, 0, 25], 
                [0, 1, 50]])
````

````python
shifted = cv2.warpAffine(image, M, (width, height))
````

__THEORY__

An affine transformation is any transformation that preserves collinearity (i.e., all points lying on a line initially still lie on a line after transformation) and ratios of distances (e.g., the midpoint of a line segment remains the midpoint after transformation).

We'll be coming back to this again.

````python
jimshow(shifted, "Shifted Down and Right")
````

__Shift up and left__

We want to shift the whole image _left_ by 50 pixels and _up_ by 90.

````python
M = np.float32([[1, 0, -50], 
                [0, 1, -90]])
````

````python
shifted = cv2.warpAffine(image, M, (width, height))
````

````python
jimshow(shifted, "Shifted Up and Left")
````

````python
M = np.float32([[1, 0, 0], 
                [0, 1, 100]])
shifted = cv2.warpAffine(image, M, (width, height))
jimshow(shifted, "Shifted down")
````

## Writing a function for our programme

````python
def translate(image, x, y):
    # Define the translation matrix and perform the translation
    M = np.float32([[1, 0, x], 
                    [0, 1, y]])
    shifted = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))

    # Return the translated image
    return shifted
````

````python
shifted = translate(image, 0, 100)
````

````python
jimshow(shifted, "Shifted down")
````
